# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
df=pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv")
df.head(12)
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python
y=df.total_profit
x=df.month_number
plt.figure(figsize=(8,5))
plt.plot(x,y,linewidth='2')

plt.title("Company profit per month",fontsize=15)
plt.xlabel("Month number",fontsize=12)
plt.ylabel("Total profit",fontsize=12)
plt.xticks(x)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
y=df.total_profit
x=df.month_number
plt.figure(figsize=(8,5))
plt.plot(x,y,color="red",linestyle='--',marker='o',markerfacecolor='k',linewidth=3)

plt.title("Company sales data of last year",fontsize=15)
plt.xlabel("Month Number",fontsize=12)
plt.ylabel("Total profit",fontsize=12)
plt.xticks(x)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.legend(labels=["Profit data of last year"],loc='lower right')
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
x=df.month_number
facecreamData=df.facecream
facewashData=df.facewash
toothpasteData=df.toothpaste
bathingsoapData=df.bathingsoap
shapooData=df.shampoo
moisturizerData=df.moisturizer
plt.figure(figsize=(8,5))
plt.plot(x,facecreamData,label='Face cream Sales Data',marker='o',linewidth=3)
plt.plot(x,facewashData,label='Face Wash Sales Data',marker='o',linewidth=3)
plt.plot(x,toothpasteData,label='ToothPaste Sales Data',marker='o',linewidth=3)
plt.plot(x,bathingsoapData,label='ToothPaste Sales Data',marker='o',linewidth=3)
plt.plot(x,shapooData,label='ToothPaste Sales Data',marker='o',linewidth=3)
plt.plot(x,moisturizerData,label='ToothPaste Sales Data',marker='o',linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(x)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

> Văn bản được thụt lề


![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python
x=df.month_number
y=df.toothpaste
plt.scatter(x,y)

plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc='upper left')
plt.xticks(x)
plt.yticks([4500,5000,5500,6000,6500,7000,7500,8000])
plt.grid(linestyle='--',linewidth=1.5)
plt.title('Tooth paste Sales data')
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
x=df.month_number
facecreamData=df.facecream
facewashData=df.facewash
plt.bar([a-0.25 for a in x],facecreamData,width=0.25,label='Face Cream sales data',align='edge')
plt.bar([a+0.25 for a in x],facewashData,width=-0.25,label='Face Wash sales data',align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title('Facewash and facecream sales data')
plt.xticks(x)
plt.grid(linewidth=1,linestyle='--')
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/07.png?raw=1)

```python
x=df.total_profit
y=[150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(x, y, label = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(y)
plt.title('Profit data')
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/08.png?raw=1)

```python
month=df.month_number
data=[df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
labels=['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.axis("equal")
plt.pie(data,labels=labels,autopct='%1.2f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/09.png?raw=1)

```python
months=df.month_number
bathingsoapData=df.bathingsoap
facewashData=df.facewash
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(months, bathingsoapData, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(months, facewashData, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')

plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/10.png?raw=1)

```python
x=df.month_number
facecreamData=df.facecream
facewashData=df.facewash
toothpasteData=df.toothpaste
bathingsoapData=df.bathingsoap
shapooData=df.shampoo
moisturizerData=df.moisturizer
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)
plt.stackplot(x, facecreamData, facewashData, toothpasteData, bathingsoapData, shapooData, moisturizerData, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
